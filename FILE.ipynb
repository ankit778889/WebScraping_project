# IMPORTING THE PACKAGES

```python
import requests
import bs4
from bs4 import BeautifulSoup
```

# CLASS FOR DATA EXTRACTION AND FILE CREATION 

```python
class webscrap: 
    def dataextraction(self,link):
        headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/42.0.2311.135 Safari/537.36 Edge/12.246"}
        data = requests.get(url=link, headers=headers)
        soup = BeautifulSoup(data.content)
        return soup
    def filemaker(self,filename,soup):
        fp=open(filename,'w',encoding='utf-8')
        fp=fp.write(str(soup))
        
```

## INITIALIZATION OF INPUT

```python
URL='https://www.imdb.com/title/tt0268093/?ref_=nv_sr_srsg_0_tt_7_nm_1_in_0_q_ramayan'
ankit=webscrap()
soup=ankit.dataextraction(URL)
filename='ankit.txt'
ankit.filemaker(filename,soup)
```

```python
temp = open('ankit.txt','r',encoding='utf-8')
c = ''
dt = []
j = 0
while(j < len(temp.readline())):
    j  = j + 1
    for i in temp.readline():
        c = c + i
        dt.append(c)
    c = ''
```
